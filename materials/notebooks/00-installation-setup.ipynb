## 3. The command-line interface

Everything in this workshop runs through the `gwmock` CLI. The six
subcommands you will meet:

| Command | What it does |
|---|---|
| `simulate` | Generate data from a YAML config (or **re-run** from metadata) |
| `config`   | List / copy the bundled example configs |
| `validate` | Check output files against the hashes recorded in metadata |
| `merge`    | Combine frame files sharing a channel (e.g. signal **+** noise) |
| `batch`    | Emit a Slurm submit script for large runs |
| `repository` | Push/pull a dataset to a Zenodo deposition |
